```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Making image
N = 256
#image = np.zeros([N,N])
guassian_image = np.zeros([N,N])
i_mid = N/2
j_mid = N/2
for i in range(N):
    for j in range(N):
        guassian_image[i][j] = np.exp(-1/400*((i-j)**2+(j)**2))
        #image[i][j] = 0.0001*((i)**2 + (j)**2)
plt.imshow(guassian_image)
#plt.imshow(image)
```

```python
f_image = np.fft.fftshift(np.fft.fft2(np.fft.fftshift(guassian_image)))
#plt.plot(abs(f_image))
plt.imshow(abs(f_image))
```
