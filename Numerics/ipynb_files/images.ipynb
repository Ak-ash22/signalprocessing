```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Making image
N = 256
#image = np.zeros([N,N])
guassian_image = np.zeros([N,N])
i_mid = N/2
j_mid = N/2
for i in range(N):
    for j in range(N):
        guassian_image[i][j] = np.exp(-((i-N/2)**2+(j-N/2)**2)/5**2)
        #image[i][j] = 0.0001*((i)**2 + (j)**2)
plt.imshow(guassian_image)
#plt.imshow(image)
```

```python
f_image = np.fft.fftshift(np.fft.fft2(np.fft.fftshift(guassian_image)))
#plt.plot(abs(f_image))
plt.imshow(abs(f_image))
```

```python
plt.imshow(np.angle(f_image))
```

## Moving gaussian by few pixels

```python
stguassian_image = np.zeros([N,N])
i_mid = N/2
j_mid = N/2
for i in range(N):
    for j in range(N):
        stguassian_image[i][j] = np.exp(-((i-N/2+20)**2+(j-N/2+20)**2)/5**2)
        #image[i][j] = 0.0001*((i)**2 + (j)**2)
plt.imshow(stguassian_image)
#plt.imshow(image)
```

```python
f_stimage = np.fft.fftshift(np.fft.fft2((stguassian_image)))
#plt.plot(abs(f_image))
plt.imshow(abs(f_stimage))
```

```python
plt.imshow(np.angle(f_stimage))
plt.colorbar()
```

## Adding 2 gaussians

```python
two_gaussian_im = guassian_image + stguassian_image
plt.imshow(two_gaussian_im)
```

```python
f_two_gaussian_im = np.fft.fftshift(np.fft.fft2(np.fft.fftshift(two_gaussian_im)))
plt.imshow(np.abs(f_two_gaussian_im))
```

```python
plt.imshow(np.angle(f_two_gaussian_im))
```

## Slit

```python
slit_image = np.zeros([N,N])
a = 20
b = 100
for i in range(N):
    for j in range(N):
        if np.abs(i-N/2)<a/2 and np.abs(j-N/2)<b/2:
            slit_image[i][j] = 1
plt.imshow(slit_image)
```

```python
f_slit_image = np.fft.fftshift(np.fft.fft2(np.fft.fftshift(slit_image)))
plt.imshow(np.abs(f_slit_image))
```

```python
plt.imshow(np.angle(f_slit_image))
```

```python
circular_slit_image = np.zeros([N,N])
for i in range(N):
    for j in range(N):
        if (i-N/2)**2 + (j-N/2)**2 < 100:
            circular_slit_image[i][j] = 1
plt.imshow(circular_slit_image)
```

```python
f_circular_slit_image = np.fft.fftshift(np.fft.fft2(np.fft.fftshift(circular_slit_image)))
plt.imshow(np.abs(f_circular_slit_image))
```

```python
plt.imshow(np.angle(f_circular_slit_image))
```
